```python
# <문제> 거스름 돈
# 최소 동전 갯수 구하기

n = 1260
count = 0
array = [500,100,50,10]
for i in array:
  count += n//i
  n = n%i

print(count)

# 화폐의 종류가 K라고 할 때, 소스코드의 시간 복잡도는 O(K)이다.
# 이 알고리즘의 시간 복잡도는 거슬러줘야 하는 금액과는 무관하며, 동전의 총 종류(K)에만 영향을 받는다.
```

```python
# <문제> 1이 될 때까지
# 1. N에서 1을 뺀다. 2. N을 K로 나눈다.
# 다음 두 과정 중 하나를 반복적으로 선택하여 수행함으로써 N이 1이 되게한다.
# N과 K가 주어졌을 때, N이 두 과정을 수행해야하는 최소 횟수 구하기

n, k = map(int, input().split())
result = 0

while True:
    target = (n//k)*k
    result += n-target
    n = target

    result += 1
    n //= k
    
    if n < k:
        break

result += n-1

print(result)
```

```python
# <문제> 곱하기 혹은 더하기
# 각 자리가 숫자로만 이루어진 문자열 S에 대해 숫자 사이에 'x' 혹은 '+'연산자를 넣어 가장 큰 수 만들기
# 단, 모든 연산은 사칙연산이 아닌 왼쪽에서부터 순서대로 이루어진다.

data = input()
result = int(data[0])

for i in range(1,len(data)):
    num = int(data[i])
    if num <= 1 or result <= 1:
        result += num
    else:
        result *= num

print(result)    
```

```python
# <문제> 모험가 길드
# 공포도가 X인 모험가는 반드시 X명 이상으로 구성한 모험가 그룹에 참여한다.
# N명의 모험가에 대한 정보가 주어졌을 때, 여행을 떠날 수 있는 그룹 수의 최댓값 구하기

n = int(input())
data = list(map(int, input().split()))
data.sort()

count = 0 # 그룹 안의 모험가 수
result = 0 # 그룹 수

for i in data:
    count += 1
    if count >= i:
        result += 1
        count = 0

print(result)
```
