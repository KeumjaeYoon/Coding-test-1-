## 거스름돈

- 1260원을 500, 100, 50, 10원으로 돌려줄 때, 거슬러 줘야할 동전의 최소 개수를 구하는 프로그램

```python
n = 1260
count = 0

for i in [500, 100, 50, 10]:
    count += n//i
    n = n%i
    
print(count)
```

- 화폐의 종류가 K라고 할 때, 소스코드의 시간 복잡도는 O(K)

## 1이 될 때까지

어떠한 수 N이 1이 될 때까지 다음의 두 과정 중 하나를 반복적으로 선택하여 수행

단, 두 번째 연산은 N이 K로 나누어 떨어질 때만 선택할 수 있다
1. N에서 1을 뺀다
2. N을 K로 나눈다

N과 K가 주어질 때, N이 1이 될 때까지 1번 혹은 2번의 과정을 수행해야 하는 최소 횟수를 구하는 프로그램

```python
n = int(input('n : '))
k = int(input('k : '))
i = 0

while True:
    if n%k == 0:
        n = n//k
        i += 1
        
    else:
        n -= 1
        i += 1
        
    if n == 1:
        break
        
print(i)
```

###### 한 번에 k에 나누어 떨어지는 n을 찾음으로써 반복 횟수를 줄여 계산 속도 향상

```python
n, k = map(int, input().split())
result = 0

while True:
    target = (n//k)*k
    result += n - target
    n = target
    if n < k:
        break
    result += 1
    n = n//k

result += n-1
print(result)
```

## 곱하기 혹은 더하기

각 자리가 숫자(0부터 9)로만 이루어진 문자열 S가 주어졌을 때, 왼쪽부터 오른쪽으로 하나씩 모든 숫자를 확인하며 숫자 사이에 'x' or '+' 연산자를 넣어 결과적으로 만들어질 수 있는 가장 큰 수를 구하는 프로그램

단, 사칙연산 기준이 아닌 모든 연산은 왼쪽에서부터 순서대로 이루어짐

```python
n = input()
first = int(n[0])

for i in range(1, len(n)):
    num = int(n[i])
    
    if num <= 1 or first <= 1:
        first += num
    
    else:
        first *= num

print(first)
```

##### 두 수 중 하나라도 1이하라면 덧셈, 두 수가 2이상이면 곱셈

## 모험가 길드

한 마을에 모험가가 N명 존재.

공포도가 X인 모험가는 반드시 X명 이상으로 구성한 모험가 그룹에 참여.

N명의 모험가에 대한 정보가 주어졌을 때, 여행을 떠날 수 있는 그룹 수의 최댓값을 구하는 프로그램

```python
n = int(input())
data = list(map(int, input().split()))
data.sort()

result = 0
count = 0

for i in data:
    count += 1
    if count >= i:
        result += 1
        count = 0

print(result)
```

## 상하좌우
A는 N x N 크기의 정사각형 공간 위에 있다. 가장 왼쪽 위 좌표는 (1,1)이며. 가장 오른쪽 아래 좌표는 (N,N)이다.

A는 상, 하, 좌, 우 방향으로 이동할 수 있으며, 시작 좌표는 항상 (1,1)이다

L : 왼쪽으로 한 칸 이동, R : 오른쪽으로 한 칸 이동

U : 위로 한 칸 이동, D : 아래로 한 칸 이동

단. A가 N x N을 벗어나는 움직임은 무시됨.

<입력 조건>
- 첫째 줄에 공간의 크기를 나타내는 N
- 둘째 줄에 A가 이동할 계획

<출력 조건>
- A가 최종적으로 도착할 지점의 좌표를 출력

```python
n = int(input())
data = list(input().split())

x, y = 1, 1

# R, L, U, D
dx = [0, 0, -1, 1]
dy = [1, -1, 0, 0]

for i in data:
    if i == 'R':
        if y == n:
            continue
        else:
            x += dx[0]
            y += dy[0]
    
    elif i == 'L':
        if y == 1:
            continue
        else:
            x += dx[1]
            y += dy[1]
    
    elif i == 'U':
        if x == 1:
            continue
        else:
            x += dx[2]
            y += dy[2]
            
    else:
        if x == n:
            continue
        else:
            x += dx[3]
            y += dy[3]
            
print(x,y)
```

## 시각
정수 N이 입력되면 00시 00분 00초부터 N시 59분 59초까지 모든 시각 중에서 3이 하나라도 포함되는 모든 경우의 수를 구하는 프로그램을 작성

<입력 조건>
- 첫째 줄에 정수 N이 입력

<출력 조건>
- 00시 00분 00초부터 N시 59분 59초까지의 모든 시각 중에서 3이 하나라도 포함되는 모든 경우의 수를 출력

가능한 모든 시각의 경우를 하나씩 세서 풀 수 있는 문제

이러한 유형은 완전 탐색(Brute Forcing) 문제 유형이라고 함

```python
n = int(input())

count = 0

for i in range(n+1):
    for j in range(60):
        for k in range(60):
            if '3' in str(i)+str(j)+str(k):
                count += 1
print(count)
```

## 왕실의 나이트

행복 왕국의 왕실 정원은 체스판과 같은 8 x 8 좌표 평면.

나이트는 L자 형태로만 이동할 수 있으며 정원 밖으로 나갈 수는 없다

- 수평으로 두 칸 이동한 후 수직으로 한 칸 이동하기
- 수직으로 두 칸 이동한 후 수평으로 한 칸 이동하기

나이트의 위치가 주어졌을 때 이동할 수 있는 경우의 수를 출력

<입력 조건>
- 첫째 줄에 8 x 8 좌표 평면상에서 현재 나이트가 위치한 곳의 좌표를 나타내는 두 문자로 구성된 문자열 입력
- ex) a1 : a열, 1행

<출력 조건>
- 첫째 줄에 나이트가 이동할 수 있는 경우의 수를 출력

```python
from string import ascii_lowercase

alpha_list = list(ascii_lowercase)
dict1 = []

for i, alpha in enumerate(alpha_list):
    dict1.append([alpha,i+1])

alpha_dict = dict(dict1)

w = input()

count = 0

dx = [2, 2, -2, -2, 1, 1, -1, -1]
dy = [1, -1, 1, -1, 2, -2, 2, -2]


x, y = alpha_dict[w[0]], int(w[1])

for i in range(8):
    new_x = x + dx[i]
    new_y = y + dy[i]
    if new_x >= 1 and new_x <= 8 and new_y >= 1 and new_y <= 8:
        count += 1
        
print(count)
```

```python
w = input()
row = int(w[1])
column = int(ord(w[0])) - int(ord('a')) + 1
print(column, row)
```

## 문자열 재정렬
알파벳 대문자와 숫자로만 구성된 문자열이 입력으로 주어질 때, 모든 알파벳을 오름차순으로 정렬하여 이어서 출력한 후 그 뒤에 모든 숫자를 더한 값을 이어서 출력.

ex) K1KA5CB7 -> ABCKK13

<입력 조건>
- 첫째 줄에 하나의 문자열 S가 주어짐

<출력 조건>
- 정답을 출력

```python
data = input()
```

```python
alpha = []
num = 0

for i in data:
    if i.isalpha() == True:
        alpha.append(i)
    else:
        num += int(i)

alpha.sort()

if num != 0:
    alpha.append(str(num))

print(''.join(alpha))
```
